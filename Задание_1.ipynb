### Открываем файл dom.txt с диска и читаем текст, приводим текст к нижнему регистру.

```python
import string
from nltk.tokenize import word_tokenize
data = open('C:/Users/sofas/Desktop/4 курс осень/Программирование/dom.txt', 'r', encoding='utf-8').read().lower()
print(data)
data.close()
```

```python
words = data.split()
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in words]
```

```python
print(words)
```

### Удаляем знаки препинания 

```python
tokens = word_tokenize(data)
words = [word for word in tokens if word.isalpha()]
```

```python
print(words)
```

### Частотный словарь

```python
from collections import Counter
counts = Counter(words)
print(counts)
```
